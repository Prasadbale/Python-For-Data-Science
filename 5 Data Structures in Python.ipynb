## Sort
- Operates in-place: It sorts the elements of the original list in ascending order without creating a new list.
- Modifies the original list: It changes the order of elements in the original list.
- Returns None: The .sort() method does not return a new sorted list; it returns None.
- Cannot be used on non-list iterables: It's specific to lists and cannot be used on other iterable types like tuples or strings.


## Sorted
- Returns a new sorted list: It sorts the elements of an iterable and returns a new list containing the sorted elements, leaving the original iterable unchanged.
- Works on various iterables: sorted() can be used on a wide range of iterable types, not just lists.
- Accepts a reverse parameter: You can specify whether to sort in ascending (default) or descending order by setting the reverse parameter to True or False.

#### Shallow Copying 

A shallow copy is a copy of an object that only creates a new reference to the original object. This means that if you change the original object, the changes will be reflected in the shallow copy.

```python
import copy

original_list = [1, 2, 3]
shallow_copy = original_list
print('Original list before changes: ',original_list)
print('Original Shallow copy: ',shallow_copy)
```

```python
shallow_copy.extend([4,5])
print('Original list after changes: ',original_list)
print('Shallow copy after changes: ',shallow_copy)
```

```python
import copy

original_list = [1, [2, 3], 4]
shallow_copy = copy.copy(original_list)

# Modifying an inner object in the shallow copy affects the original
shallow_copy[1][0] = 99
print(original_list)  # Output: [1, [99, 3], 4]
```

## Deep copy
A deep copy is a copy of an object that creates a new object and copies all of the data from the original object into the new object. This means that if you change the original object, the changes will not be reflected in the deep copy.
To create a deep copy in Python, you can use the deepcopy() function.

```python
import copy

original_list = [1, 2, 3]
shallow_copy = copy.deepcopy(original_list)
print('Original list before changes: ',original_list)
print('Original Shallow copy: ',shallow_copy)
```

```python
shallow_copy.extend([4,5])
print('Original list after changes: ',original_list)
print('Shallow copy after changes: ',shallow_copy)
```

<h1 style="color:Brown"> Dictionaries </h1>

- A dictionary stores element as keys and values pairs.
- The key is like an index, it's is always unique and immutable. 
- The values are the objects that contain information.
- Values are accessed using their keys.
- Each key is followed by a value separated by a colon. 
- The values can be immutable, mutable, and duplicates. 
- Each key and value pair is separated by a comma enclosed inside curly brackets.

```python
# Creating a dictionary

d = {"India" : "INR", "USA" : "USD", "France" : "Euros"}

# Access value using keys

d["USA"]
```

```python
# Replace the value for a key in a dictionary

d["USA"] = "$"

d
```

```python
# Insert new key value pair into a dictionary

d["Japan"] = "Yen"

d
```

```python
# Deleting a key value pair

del d["France"]

d
```

```python
# Sorting a dictionary

sorted(d)
```

```python
# values() Method

d = {"India" : "INR", "USA" : "USD", "France" : "Euros"}

d.values()
```

```python
# Keys() Method

d.keys()
```

```python
# Update() Method

d.update({'USA':'$'})
d
```
